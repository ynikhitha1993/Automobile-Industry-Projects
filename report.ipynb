```python
# Combined line and bar plot
fig, ax1 = plt.subplots(figsize=(10, 6))

# Bar plot
ax1.bar(data['Quarter'], data['Value'], color='lightgray', label='Bar')

# Line plot
ax2 = ax1.twinx()
sns.lineplot(data=data, x='Quarter', y='Value', marker='o', ax=ax2, label='Line', color='blue')

ax1.set_xlabel('Quarter')
ax1.set_ylabel('Bar Values', color='gray')
ax2.set_ylabel('Line Values', color='blue')
ax1.set_title('Combined Bar and Line Plot')

fig.legend(loc='upper right')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Create a DataFrame with the data
data = {
    'Category': ['Automotive', 'Automotive', 'Automotive', 'Automotive', 'Automotive',
                 'BMW', 'Eliminations', 'Financial Services', 'Group', 'Group', 'Group'],
    'Subcategory': ['Deliveries', 'EBIT margin', 'Free cash flow', 'All-electric cars in deliveries', 'Revenues',
                    'BMW Deliveries', 'Revenues', 'Revenues', 'Earnings per share', 'Profit before tax', 'Revenues'],
    '2023': [2429728, 13343.6, 3141, 12.6, 62898,
             1071326, -8374, 17621, 9.7, 9351, 74072],
    '2024': [2426552, 10796.6, 2289, 15.7, 63009,
             1096423, -10586, 19267, 8.57, 8023, 73558]
}

df = pd.DataFrame(data)

# Step 2: Plot the data
plt.figure(figsize=(12, 8))

# Plot lines for 2023 and 2024
for category in df['Category'].unique():
    subset = df[df['Category'] == category]
    plt.plot(subset['Subcategory'], subset['2023'], marker='o', label=f'{category} 2023')
    plt.plot(subset['Subcategory'], subset['2024'], marker='o', linestyle='--', label=f'{category} 2024')

# Add labels, title, and legend
plt.title('Comparison of Subcategories for 1 January to 30 June 2023 and 2024')
plt.xlabel('Subcategory')
plt.ylabel('Values')
plt.xticks(rotation=45)
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))

# Display the plot
plt.grid(True)
plt.show()
```
